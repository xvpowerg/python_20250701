```python
h = input("請輸入身高單位公分:")
w = input("請輸入體重單位公斤:")
h = int(h)
w = int(w)
bmi = w / (h/100) ** 2
print(bmi)
print(round(bmi,5) )#round 取道小數點第n位
if bmi < 18.5:
    print("太輕")
elif bmi>= 18.5 and bmi <=25:
     print("正常")
elif bmi >25 and bmi <= 30:
    print("過重")
else:
    print("胖胖")
```

```python
n = "y"

while n == "y":
    a = int(input("請輸入第一個整數"))
    b = int(input("請輸入第二個整數"))
    ans = a * b
    print(f"{a}*{b}={ans}")
    n = input("是否繼續是按y")
```

```python
# 顯示2 ~ 15 
n = 5
for i in range(2,16):
    print(i,end = " ")
    if i == n:
        break#停止迴圈
```

```python
for i in range(2,16):
    if i == 5 or i == 9:
        continue
    print(i,end=" ")
```
