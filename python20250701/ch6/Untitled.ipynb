## 作業 文字接龍

```python
txt = input("請輸入第一個字串")
strList = []
strList.append(txt)
msg = txt
count = 0
inputMsg = "請輸入-"+txt[-1]+"-開始的字串"
while count < 5:
    tmpTxt = input(inputMsg)
    if tmpTxt[0] == txt[-1]:
        txt = tmpTxt
        strList.append(txt)
        msg = "-".join(strList)
        inputMsg = "請輸入-"+txt[-1]+"-開始的字串"
    else:
         inputMsg = "請再次輸入-"+txt[-1]+"-開始的字串"
         count += 1
    print("上一個字串是:",msg)        
```

## list 項目擴展資料
+ extend 擴展資料：
  + 如果引號括起來的字串有 [ ] 將會拆解成好幾個資料。
  + 可以把其他 list 加入到這個 list 內，擴展多個欄位。

```python
myList = []
print(len(myList))
myList.append("Ken")
myList.append("Vivin")
print("myList len:",len(myList))
print(myList)

newList = ["Joy","Lucy","Iris","Sean"]
print(len(newList))
myList.append(newList)
print("myList len:",len(myList))
print(myList)
```

```python
myList2 = [12,35,67]
print(len(myList2))
newList2 = [102,256,720,560]
myList2.extend(newList2)
print(len(myList2))
print(myList2)
```

```python
myList = ["a","b","c"]
myList.insert(2,"X")
print(myList)
myList.insert(0,"Y")
print(myList)
```

### pop

如果 pop( ) 內沒有參數，則移除最後一筆資料。

pop() 內若有參數，則移除指定位置的資料。

```python
list2 = []
list2.append("Ken")
list2.append("Vivin")
list2.append("Lindy")
print(list2)
v1 = list2.pop()
print(list2)
print("v1:",v1)

v2 = list2.pop(0)
print(list2)
print("v2:",v2)
v3 = list2.pop()
print(list2)
print("v3:",v3)
list2.pop()
```

```python
list3 = ["Ken","Vivin","Lindy","Lucy"]
print(list3)
while len(list3) != 0:
    v1 = list3.pop()
    print(v1)
```

### Del
+ list 可透過 del 方式移除指定位置資料。
+ del list[ ] 必須指定範圍：
  + del list[n] 代表移除第 n 位索引值資料。
  + del list[:m] 代表移除第 m 位索引值之前 ( 不包含 m ) 資料。
  + del list[n:m] 代表移除第 n 到 m 位索引值之前 ( 不包含 m ) 資料。

```python
list2 = ["A","B","C","D","E"]
print(list2)
del list2[2]
print(list2)
del list2[1:3]
print(list2)
```

## Dictionary

資料結構是由 key:value 所組成。

key 不能夠重複，否則會被後面的結果蓋過去。

可輸入 key 找尋您要找出來的值。

如果輸入的 key 不存在，那就會出現錯誤訊息。

key 只能是 字串 數字 tuple

```python
dict1 = {"a":100,"b":200,"c":300}
print(dict1["b"])
print(dict1["c"])
print(dict1["x"])#KeyError 因為x的key不存在所以出錯
```

```python
myList = ["Ken","Vivin","Lucy"]
for name in myList:
    print(name)
```

```python
dict1 = {"a":100,"b":200,"c":300}
for k in dict1:
    print(k,":",dict1[k])
```

```python
dict2 = {"ken":100,"vivin":95,"lucy":72}
print(dict2)
dict2["vivin"] = 77#key在更新
print(dict2)
dict2["iris"] = 88 #key不在變新增
print(dict2)
```
