# 比較運算子

```python
a1 = 10
a2 = 10

print(a1,">",a2,a1 > a2)
print(a1,"<",a2,a1 < a2)
print(a1 ,">=",a2,a1 >= a2)
print(a1,"<=",a2,a1 <= a2)

print(a1,"==",a2,a1 == a2)
print(a1,"!=",a2,a1 != a2) #不等於
```
